```python
#%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple line graph
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
ax3.plot(np.random.standard_normal(50).cumsum(), color= 'black', linestyle = 'dashed')
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
ax1.hist(np.random.standard_normal(100), bins = 20, color = 'black', alpha = 0.3)
ax4.scatter(np.random.standard_normal(30), np.random.standard_normal(30), color = 'green')
ax4.scatter(np.random.standard_normal(30)+3, np.random.standard_normal(30)+2 , color = 'red')
```

```python
ax = plt.figure().add_subplot()
ax.plot(np.random.standard_normal(30).cumsum(), color = 'black', linestyle = 'dashed',marker = 'o');
```

```python
fig , axes = plt.subplots(2, 2, sharex = True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500), color = 'black', bins = 50, alpha = 0.5)
fig.subplots_adjust(wspace = 0, hspace = 0)
```

```python
fig2 , ax  = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum(), label = 'line 1')
ax.set_title("coin prices")
ax.set_xlabel("years----->")
ax.set_ylabel("in crores---->")
#Setting legends
ax.plot(np.random.standard_normal(1000).cumsum(), color = 'black', linestyle = 'dotted', label = 'line2')
ax.plot(np.random.standard_normal(1000).cumsum(), color = 'black', linestyle = 'dashed', label = 'line3')
ax.legend()
ax.text(642, 10, "Hello world!",
        family="monospace", fontsize=10)
```
