```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Training Dataset
dataset_train = pd.read_csv("train.csv")
dataset_train["Age"].fillna(value=29.699118,axis = 0, inplace = True)


x_train = dataset_train.loc[:,["Pclass","Sex","Age","SibSp","Parch","Fare"]].values
y_train = dataset_train.iloc[:,1].values

#Testing Data
dataset_test = pd.read_csv("test.csv")
dataset_test["Age"].fillna(value=29.699118,axis = 0, inplace = True)
x_test = dataset_test.loc[:,["Pclass","Sex","Age","SibSp","Parch","Fare"]].values
y_test = dataset_test.iloc[:,1].values
```

```python
x_train[0]
```

```python
# OneHot Encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import  OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[1])], remainder = 'passthrough')
#ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[-1])], remainder = 'passthrough')

x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
x_train[0]
```

```python
for i in x_train:
    print(i)
```

```python
#Standardising the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
```

```python
from sklearn.tree import  DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regsr = RandomForestRegressor(n_estimators=10, random_state=0)
regsr.fit(x_train,y_train)
y_pred2 = regsr.predict(x_test)
```

```python
#Making confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

y_pred2list = []
for x in regsr.predict(x_test):
    if x > 0.2:
        y_pred2list.append(1)
    else:
        y_pred2list.append(0)
cm2 = confusion_matrix(y_test, np.array(y_pred2list))
print(cm)
print(accuracy_score(y_test,y_pred)*100)
print(accuracy_score(y_test,np.array(y_pred2list))*100)
```
