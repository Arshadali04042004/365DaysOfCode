```python
# loc an iloc
# loc is used where we have to select a range of Data by providing name of the rows and cols .
# iloc is used where we have to select a range of Data by providing index of the rows and cols.

newdf3 = pd.read_csv("students_list.csv")
data_frame_names = newdf3.loc[:,"name"]
data_frame_marks = newdf3.loc[:,"makrs"]
```

```python
data_frame_names
```

```python
data_frame_marks
```

```python
#Using iloc
names = newdf3.iloc[0:5,0:1]
marks = newdf3.iloc[:,-1]
```

```python
marks
```

```python
# To get the result as array use .values
names = newdf3.iloc[0:5,0:1].values
marks = newdf3.iloc[:,-1].values
print(names)
print(marks)
```

```python
# To delete a row or column use drop
newdf3_drop1 = newdf3.drop(0,axis=0)   #axis = 0 means it will select rows and 
newdf3_drop2 = newdf3.drop("makrs",axis =1)   #axis = 1 menas it will select columns
print(newdf3_drop1)
print()
print(newdf3_drop2)
```

```python
newdf3   # use inplace = True to make the changes in the original Dataframe
```
