```python
#importing data preprocessing lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset
df = pd.read_csv('Position_Salaries.csv',sep = ',')
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
#printing x and y values
print(x)
print()
print(y)
```

```python
#Training the model using linear model
from sklearn.linear_model import LinearRegression
LiReg = LinearRegression()
LiReg.fit(x,y)
```

```python
#Training the model using polynomial reg
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
#ploting graph of Linear Regression 
plt.scatter(x,y,color = 'red')
plt.plot(x,LiReg.predict(x),color = 'blue')
plt.xlabel('Positional Level')
plt.ylabel('Salary')
plt.show()
```

```python
#ploting graph of Polynomial Regression 
plt.scatter(x,y,color = 'red')
plt.plot(x,lin_reg2.predict(x_poly),color = 'blue')
plt.xlabel('Positional Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Now predicting the new salary with linear reg model first
LiReg.predict([[6.5]])
```

```python
#Now predicting the new salary with poly nomial model as result of linear model is not that accurate
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
